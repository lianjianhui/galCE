```python
from astropy.io import fits
from astropy.io import ascii
from pylab import *

#read in the output of galce.py for for a model with exponentially declining gas accretion history.
track = fits.open('tracks/model3.fits')

info = track[1].data #basic information, gas and stellar mass, sfr etc.
abund = track[2].data #element abundance
print (info['stellarmass'][-1])

plt.figure(figsize=(11,5))
plt.subplot(121)
plt.plot(info['t'],info['gasmass'])
plt.xlabel('Evolutioni time (Gyr)',fontsize=20)
plt.ylabel(r'${\rm log(M_{gas}/M_{\odot})}$',fontsize=20)
plt.tick_params(which='both',width=1.3)
plt.tick_params(which='major',length=8,labelsize=18,pad=5,direction="in",top=True,right=True)
plt.minorticks_on()
plt.tick_params(which='minor',length=5,direction="in",top=True,right=True)
plt.locator_params(nbins=5)

plt.subplot(122)
plt.plot(info['t'],info['sfr'])
plt.xlabel('Evolutioni time (Gyr)',fontsize=20)
plt.ylabel(r'${\rm log(SFR/M_{\odot}yr^{-1})}$',fontsize=20)
plt.tick_params(which='both',width=1.3)
plt.tick_params(which='major',length=8,labelsize=18,pad=5,direction="in",top=True,right=True)
plt.minorticks_on()
plt.tick_params(which='minor',length=5,direction="in",top=True,right=True)
plt.locator_params(nbins=5)

plt.tight_layout()
plt.show()
```

```python
#check the model track in [Fe/H]-[alpha/Fe] and [Fe/H]-[Eu/Fe] plane. 
feh1 = abund[:,25]
sih1 = abund[:,13]
oh1 = abund[:,7]
sh1 = abund[:,15]
sife1 = sih1-feh1
ofe1 = oh1-feh1
euh  = abund[:,62]
eufe = euh-feh1

plt.figure(figsize=(16,5))
plt.subplot(131)
plt.plot(feh1,ofe1,'-')
plt.xlim(-3,0.6)
plt.ylim(-0.4,1)
plt.xlabel('[Fe/H]',fontsize=20)
plt.ylabel('[O/Fe]',fontsize=20)
plt.tick_params(which='both',width=1.3)
plt.tick_params(which='major',length=8,labelsize=18,pad=5,direction="in",top=True,right=True)
plt.minorticks_on()
plt.tick_params(which='minor',length=5,direction="in",top=True,right=True)
plt.locator_params(nbins=5)

plt.subplot(132)
plt.plot(feh1,sife1,'-')
plt.xlim(-3,0.6)
plt.ylim(-0.4,1)
plt.tick_params(which='both',width=1.3)
plt.tick_params(which='major',length=8,labelsize=18,pad=5,direction="in",top=True,right=True)
plt.minorticks_on()
plt.tick_params(which='minor',length=5,direction="in",top=True,right=True)
plt.locator_params(nbins=5)
plt.xlabel('[Fe/H]',fontsize=20)
plt.ylabel('[Si/Fe]',fontsize=20)

plt.subplot(133)
plt.plot(feh1,eufe,'-')
plt.xlim(-3,0.6)
plt.ylim(-0.4,1)
plt.tick_params(which='both',width=1.3)
plt.tick_params(which='major',length=8,labelsize=18,pad=5,direction="in",top=True,right=True)
plt.minorticks_on()
plt.tick_params(which='minor',length=5,direction="in",top=True,right=True)
plt.locator_params(nbins=5)
plt.xlabel('[Fe/H]',fontsize=20)
plt.ylabel('[Eu/Fe]',fontsize=20)

plt.tight_layout()
plt.show()
```
